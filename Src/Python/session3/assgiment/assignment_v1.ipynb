# Write a Python program to remove an item from a given set

## Example
my_set: {1, 2, 3, 4, 5}

Output: {1, 2, 4, 5}

```python
my_set = {1, 2, 3, 4, 5}
item=input("Enter an item to remove from the set: ")
item = int(item)
if item in my_set:
    my_set.remove(item)
    print(f"Set after removing {item}: {my_set}")
else:
    print(f"Item {item} not found in the set.")
```

# Write a Python program to removes multiple items from a given set

## Example
my_set: {1, 2, 3, 4, 5}

items to be removed: {1, 4}

Output: {2, 3, 5}

```python
my_set={1,2,3,4,5}
items_to_remove = {2, 4, 6}
result_set = my_set - items_to_remove
print("Set after removing items:", result_set)
```

# find the factorial of a number 
## Example
(factorial of 5) = 5 * 4 * 3 * 2 * 1 = 120

```python
factorial = 1
number = int(input("Enter a number: "))
for i in range(1, number + 1):
    factorial *= i
print(f"The factorial of {number} is {factorial}")
```

# Create a function that Calculates the Sum of Even Numbers in a List
## Example
numbers: [1, 2, 3, 4, 5, 6]

Output: 12

```python
def sum_of_even_list(numbers):
    return sum(x for x in numbers if x % 2 == 0)
numbers = [1, 2, 3, 4, 5, 6]
result = sum_of_even_list(numbers)
print("Sum of even numbers in the list:", result)
```

# create a function that calculates the median of a list of numbers (a median is the middle element in sorted list of numbers)

## Example
numbers: [5, 2, 7, 3, 1]

Output: 3

```python
def median(numbers):
    sorted_numbers = sorted(numbers)
    mid = len(sorted_numbers) // 2
    return sorted_numbers[mid]

numbers = [5, 2, 7, 3, 1]
print(median(numbers))
```
