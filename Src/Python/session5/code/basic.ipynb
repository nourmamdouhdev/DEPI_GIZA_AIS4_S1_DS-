```python
class clec :
    
    def __init__(self, num1 :int, num2 : int):
        self.num1 = num1
        self.num2 = num2
        
        
    def sum (self):
        return self.num1 + self.num2
        
    def sub (self):
        return self.num1 - self.num2
```

```python
c1 = clec(5,3)
c1.sum()
```

```python
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def move(self):
        pass
    
class Dog(Animal):
    def move(self):
        return "Dog is running"
    
class cat (Animal):
    def move(self):
        return "cat is jumping"
    
```

```python
d1 = Dog()
d1.move()
cat1 = cat()
cat1.move()
```
